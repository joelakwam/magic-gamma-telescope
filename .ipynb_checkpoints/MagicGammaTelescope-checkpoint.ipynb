# Major Atmospheric Gamma Imaging Cherenkov (MAGIC) Telescope Project. 

```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib as mtplb
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
dataset_column_names = ['fLength', 'fWidth', 'fSize', 'fConc', 'fConc1', 'fAsym', 'fM3Long', 'fM3Trans', 'fAlpha', 'fDist', 'class']
dataset = pd.read_csv('magic04.data', names = dataset_column_names)
```

```python
dataset.head()
```

```python
dataset['class'].unique()
```

```python
# - Convert class names to integers.
dataset['class'] = (dataset['class'] == 'g').astype(int)
```

```python
dataset['class'].unique()
```

```python
for label in dataset_column_names[:-1]:
    plt.hist(dataset[dataset['class'] == 1][label], color = 'blue', label = 'Gamma')
    plt.hist(dataset[dataset['class'] == 0][label], color = 'yellow', label = 'Hadron', alpha = 0.5)
    plt.title(label)
    plt.ylabel('Probability')
    plt.xlabel(label)
    plt.legend()
    plt.show()
```

```python
target = dataset['class']
features = dataset.drop(['class'], axis = 1)
```

# Split Dataset Into Train, Validation & Testing. 

```python
from sklearn.cross_validation import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size = 0.2, random_state = 0)
```

## 1. K-Nearest Neighbors

```python
knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
knn_predictions = knn.predict(X_test)
print(accuracy_score(knn_predictions, y_test))
```

## 2. Logistic Regression

```python
logistic_regressor = LogisticRegression()
logistic_regressor.fit(X_train, y_train)
logistic_regressor_predictions = logistic_regressor.predict(X_test)
print(accuracy_score(logistic_regressor_predictions, y_test))
```

## 3. Support Vector Machines

```python
svm = SVC()
svm.fit(X_train, y_train)
svm_predictions = svm.predict(X_test)
print(accuracy_score(svm_predictions, y_test))
```

## 4. Naive Bayes

```python
naive_bayes = GaussianNB()
naive_bayes.fit(X_train, y_train)
naive_bayes_predictions = naive_bayes.predict(X_test)
print(accuracy_score(naive_bayes_predictions, y_test))
```
